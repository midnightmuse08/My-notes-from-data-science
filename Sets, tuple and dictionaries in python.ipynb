```python
def sorted_by_age(data):
    result = []
    empty_list = [i[1] for i in data]
    empty_list.sort()
    for i in empty_list :
        for el in data :
            if i == el[1] :
                result.append(el)

    return result


# :
people = [("", 30), ("", 22), ("", 25)]
# : [('', 22), ('', 25), ('', 30)]
```

```python
data= [("", 30), ("", 22), ("", 25)]
sorted_by_age(data)
```

```python
def word_frequency(text):
    #   
    result = {}
    splitted_text = text.split()
    for words in splitted_text:
        result[words] = splitted_text.count(words)
        
    print(" :", result)

# :
sentence = "python is great and python is easy"
# : {'python': 2, 'is': 2, 'great': 1, 'and': 1, 'easy': 1}
```

```python
def average_scores(grades_dict):
    #   
    result = {}
    for name, grades in grades_dict.items():
        if grades:
            result[name]= sum(grades) / len(grades)
        else:
            result[name]= 0

    print(":", result)



# :
grades = {
    'Alice': [80, 90, 100],
    'Bob': [70, 75],
    'Eve': [100, 100, 100]
}

# : {'Alice': 90.0, 'Bob': 72.5, 'Eve': 100.0}
```

```python
grades = {
    'Alice': [80, 90, 100],
    'Bob': [70, 75],
    'Eve': [100, 100, 100]
}
average_scores(grades)
```

```python
def reverse_dict(original):
    #   
    result = {}
    for name , department in original.items():
        result[department] = name 
    return result

# :
employees = {
    'Alice': 'HR',
    'Bob': 'IT',
    'Charlie': 'IT',
    'Diana': 'HR',
}

# :
# {
#     'HR': ['Alice', 'Diana'],
#     'IT': ['Bob', 'Charlie']
# }
```

```python
def filter_by_price(products, threshold):
    #   
    result = {}
    for items, price in products.items():
        if price < threshold:
            result[items] = price 
    return result

# :
items = {
    'apple': 1.2,
    'banana': 0.8,
    'avocado': 2.5,
    'mango': 1.5
}

# filter_by_price(items, 1.5)  {'apple': 1.2, 'banana': 0.8}
```
