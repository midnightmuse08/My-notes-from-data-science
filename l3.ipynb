**Math library in python**

```python
import math
```

```python
# Constants
print(f"Mathematical pi : {math.pi}")
print(f"Mathematical e : {math.e}")
```

```python
# Functions
print(math.fabs(-5))     
print(math.pow(2, 3))    
print(math.sqrt(25))     
```

```python
# Trigonometry
print(math.sin(math.pi / 2))  
print(math.cos(0))           
print(math.tan(math.pi / 4))  

print(math.degrees(math.pi))   
print(math.radians(180))        
```

```python
# Logarithm
print(math.exp(1))        
print(math.log(10))       
print(math.log10(1000))    
```

```python
# Rounding
print(math.floor(3.9))   
print(math.ceil(3.1))    
```
