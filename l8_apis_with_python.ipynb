```python
import requests

url = "https://itunes.apple.com/search?term=Blackpink&media=music&attribute=artistTerm&limit=3"

response = requests.get(url)
response
```

| Status Code                                | Category        | Meaning                                                           | Example Scenario                                  |
| ------------------------------------------ | --------------- | ----------------------------------------------------------------- | ------------------------------------------------- |
| **200 OK**                                 | ✅ Success       | The request worked, and the server returned the expected result.  | A GET request for weather data returns JSON.      |
| **201 Created**                            | ✅ Success       | The request created something new on the server.                  | POST request adds a new blog post.                |
| **204 No Content**                         | ✅ Success       | Request was successful, but there’s no data in the response body. | Deleting a record returns no body.                |
| **301 Moved Permanently**                  | 🔄 Redirection  | Resource has a new URL permanently.                               | Website moved to a new domain.                    |
| **302 Found** / **307 Temporary Redirect** | 🔄 Redirection  | Resource temporarily at a different URL.                          | API endpoint temporarily changes.                 |
| **400 Bad Request**                        | ⚠️ Client Error | The request was invalid or missing parameters.                    | Forgetting a required query parameter.            |
| **401 Unauthorized**                       | ⚠️ Client Error | Authentication required but not provided or invalid.              | Calling an API without the correct API key.       |
| **403 Forbidden**                          | ⚠️ Client Error | Server understood request but refuses to authorize it.            | Trying to access private data without permission. |
| **404 Not Found**                          | ⚠️ Client Error | The server can’t find the resource requested.                     | Typo in the endpoint URL.                         |
| **405 Method Not Allowed**                 | ⚠️ Client Error | Method not allowed for that endpoint.                             | Sending a POST to a GET-only endpoint.            |
| **429 Too Many Requests**                  | ⚠️ Client Error | The client sent too many requests in a short time (rate limit).   | Hitting API rate limits.                          |
| **500 Internal Server Error**              | ❌ Server Error  | Something went wrong on the server.                               | Server bug or crash.                              |
| **502 Bad Gateway**                        | ❌ Server Error  | Invalid response from an upstream server.                         | API gateway forwards a bad response.              |
| **503 Service Unavailable**                | ❌ Server Error  | Server is overloaded or down for maintenance.                     | API temporarily unavailable.                      |
| **504 Gateway Timeout**                    | ❌ Server Error  | Upstream server took too long to respond.                         | API request takes too long and times out.         |

```python
# Cat API example
res_cat = requests.get(url = "https://api.thecatapi.com/v1/images/search").json()
res_cat
```

```python
# Cat API example
url = "https://api.thecatapi.com/v1/images/search"
param = {
    'limit' : 10,
    'has_breed' : 1
}
res_cat = requests.get(url = url, params = param).json()
res_cat
```
