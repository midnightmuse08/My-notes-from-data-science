```python
class animal :
    def __init__(self, type, number_of_legs, number_of_hands, is_domestic):
        self.type = type
        self.number_of_legs = number_of_legs
        self.number_of_hands = number_of_hands
        self.is_domestic = is_domestic
    
    def sound(self):
        if self.type == "dog":
            return "bark"
        elif self.type == "cat":
            return "meow"
        elif self.type == "cow":
            return "moo"
        else:
            return "unknown sound"

    def describe(self):
        description = f"This is a {self.type} with {self.number_of_legs} legs and {self.number_of_hands} hands."
        if self.is_domestic:
            description += " It is a domestic animal."
        else:
            description += " It is a wild animal."
        return description
```

```python
# Inherited class for mammals
class mammal(animal):
    def __init__(self, type, number_of_legs, number_of_hands, is_domestic, fur_color):
        super().__init__(type, number_of_legs, number_of_hands, is_domestic)
        self.fur_color = fur_color
    
    def describe(self):
        base_description = super().describe()
        return f"{base_description} It has {self.fur_color} fur."
```

```python
# Example usage
lion = mammal("lion", 4, 0, False, "golden")
print(lion.describe())
```

```python
lion.sound()
```

```python
# Example usage 2
dog = mammal("dog", 4, 0, True, "brown")
print(dog.sound())
```

**Employee class**

```python
### Your code goes here
class employee:
    def __init__(self, name,job_type, experience, age, skin_tone, fulltime_worker ):
        self.name = name
        self.job_type = job_type
        self.experience = experience
        self.age = age
        self.skin_tone = skin_tone
        self.fulltime_worker= fulltime_worker


    def describe(self):
        description = f"this is {self.name}, he is a {self.job_type} and he has {self.experience} years experience. Also he is {self.age} years old and {self.skin_tone}. "
        if self.fulltime_worker == "full time worker":
            description += "he is a full time worker"
        else:
            description += "Unfortunately, he is not full time worker"
        return description


    
```

```python
Samir = employee("Samir", "assistent", 2, 23, "brown", "full bo time worker")
print(Samir.describe())
```

```python
class manager(employee):
    def __init__(self, name, job_type, experience, age, skin_tone, fulltime_worker, eyecolor):
        super().__init__(name, job_type, experience, age, skin_tone, fulltime_worker)
        self.eyecolor= eyecolor

    def describe(self):
        maindescribe = super().describe()
        return f"{maindescribe}. Also, he has {self.eyecolor} eyes"
```

```python
Bob= manager("Bob", "manager", 3, 34, "white", "full time worker", "bright blue")
print(Bob.describe())
```

```python
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt

class LabeledRectangle(Rectangle):
    def __init__(self, xy, width, height, label, **kwargs):
        super().__init__(xy, width, height, **kwargs)
        self.label = label

    def draw_label(self, ax):
        x, y = self.get_xy()
        ax.text(x + self.get_width()/2, y + self.get_height()/2, 
                self.label, color='white')

# Use it in a plot
fig, ax = plt.subplots()
rect = LabeledRectangle((0.8, 0.3), 0.4, 0.2, "Im a child!", edgecolor='black', facecolor='green')
ax.add_patch(rect)
rect.draw_label(ax)
plt.show()
```

```python
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt

class cloneRectangle(Rectangle):
    def __init__(self, xy, width, height, text,**kwargs):
        super().__init__(xy, width, height, **kwargs)
        self.text = text

    def draw_text(self, ax):
        x, y = self.get_xy()
        ax.text(x + self.get_width(),  y + self.get_height(),
                self.text, color= "pink")
        

fig, ax = plt.subplots()
rect = cloneRectangle((0.2, 0.3), 0.4, 0.2, "i love haikyy", edgecolor= "black", facecolor= "orange")
ax.add_patch(rect)
rect.draw_text(ax)
plt.show()
```
