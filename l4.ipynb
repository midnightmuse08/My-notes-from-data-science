**How do we create dataframes**

```python
# First technique
data = [
    ['Alice', 24],  # Rows
    ['Bob', 27],
    ['Charlie', 22]
]

df = pd.DataFrame(data, columns = ['name', 'age'])
df
```

```python
# Second technique
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [24, 27, 22]
}

df = pd.DataFrame(data)
df
```

```python
# Third techniques
s1 = pd.Series(['Alice', 24], index=['Name', 'Age'])
s2 = pd.Series(['Bob', 27], index=['Name', 'Age'])

df = pd.DataFrame([s1, s2])
df
```
