```python
class car : 
    def __init__(self, model, year, horsepower, color):
        self.model = model
        self.year = year
        self.horsepower = horsepower
        self.color = color
    
    def display_info(self):
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")
        print(f"Horsepower: {self.horsepower}")
        print(f"Color: {self.color}")

    def start_engine(self):
        print(f"The engine of the {self.model} is now running.")

    def stop_engine(self):
        print(f"The engine of the {self.model} has been turned off.") 
```

```python
# Example usage
my_car = car("Toyota Camry", 2020, 203, "Blue")
my_car.display_info()
```

```python
my_car.start_engine()
```

```python
my_car.stop_engine()
```

```python
# Real-life example 
class smartphone:
    def __init__(self, brand, model, storage, color):
        self.brand = brand
        self.model = model
        self.storage = storage
        self.color = color
    
    def display_info(self):
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Storage: {self.storage}GB")
        print(f"Color: {self.color}")

    def make_call(self, number):
        print(f"Calling {number} from {self.brand} {self.model}...")

    def send_message(self, number, message):
        print(f"Sending message to {number}: '{message}'")
```

```python
# Sample usage
iphone = smartphone(brand = "Iphone", model = "Iphone 17", storage = 512, color = "Black")
iphone.display_info()
```

```python
iphone.make_call("+998901111111")
```

```python
iphone.send_message("+998901111111", "Assalomu alaykum")
```

**PyGame**

```python
import turtle

class MyTurtle:
    def __init__(self, color, shape):
        self.t = turtle.Turtle()
        self.t.color(color)
        self.t.shape(shape)

    def move_forward(self, distance):
        self.t.forward(distance)

    def turn(self, angle):
        self.t.left(angle)

red_turtle = MyTurtle("red", "turtle")
blue_turtle = MyTurtle("blue", "circle")

red_turtle.move_forward(100)
blue_turtle.turn(90)
blue_turtle.move_forward(100)

turtle.done()
```
