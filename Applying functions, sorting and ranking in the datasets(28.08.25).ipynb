```python
# Converting age given in days to years
df_cardio['age'] = df_cardio['age'].apply(lambda x : x / 365)
df_cardio
```

```python
# Converting sex column to numerical values using function
def conv(sex) :
    if sex == 'male' : 
        return 1
    else : 
        return 0
    
df_tit['Sex'] = df_tit['Sex'].apply(conv)
df_tit.head()
```
